# **1. 사용자 건강 설문 데이터 기반 추천**
영양소 이름과 동의어를 기준으로 텍스트에서 포함 여부를 검사하고, 포함된 영양소 비율을 유사도로 계산하여 상위 제품을 추천

```python
import pandas as pd
from IPython.display import display

# 상위 3개의 영양성분 이름만 추출
top_3_nutrients_name = [nutrient[0] for nutrient in top_3_nutrients]

# 동의어 사전: 영양소명과 그 동의어
synonyms = {
    "비타민 C": ["비타민C", "비타민 C"],
    "비타민 B군": ["비타민B", "비타민 B군"],
    "오메가-3 지방산": ["오메가3", "오메가-3 지방산"],
    "마그네슘": ["마그네슘", "Mg"],
    "칼슘": ["칼슘", "Ca"],
    "단백질": ["단백질", "Protein"]
}

# 영양소가 포함되어 있는지 검사하는 함수
def nutrient_in_text(text, nutrient):
    """동의어 중 하나라도 text에 포함되면 True 반환"""
    return any(syn in text for syn in synonyms.get(nutrient, [nutrient]))

# 포함 여부 및 포함된 영양소 반환 함수
def calculate_similarity_and_extract(rawmtrl_nm, nutrients):
    included_nutrients = [nutrient for nutrient in nutrients if nutrient_in_text(rawmtrl_nm, nutrient)]
    similarity_score = len(included_nutrients) / len(nutrients)  # 포함된 영양소 비율
    return similarity_score, included_nutrients

# 유사도와 포함된 영양소 계산
df_items[["similarity", "included_nutrients"]] = df_items["RAWMTRL_NM"].apply(
    lambda x: pd.Series(calculate_similarity_and_extract(x, top_3_nutrients_name))
)

# 유사도 상위 5개 제품만 출력
df_items_sorted = df_items.sort_values(by="similarity", ascending=False)
top_5_recommendations = df_items_sorted.head(5)

# 결과 출력
print("AI 기반 영양소 포함 제품 추천 (상위 5개):")
display(top_5_recommendations[["PRDLST_NM", "RAWMTRL_NM", "similarity", "included_nutrients"]])
```

# **2. 증상 기반 영양제 추천**
FAISS를 사용해 질문과 가장 관련 있는 건강 보조제를 검색하고, GPT-3.5 Turbo를 통해 검색 결과를 기반으로 사용자 질문에 답변을 생성하는 RAG(Retrieval-Augmented Generation) 시스템

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import faiss
import openai

# OpenAI API 키 설정
openai.api_key = "openai key"

# 데이터프레임 준비: df_itmes

# 1. SentenceTransformer로 텍스트 임베딩 생성
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
embedder = SentenceTransformer(model_name)
text_data = df_items.apply(
    lambda row: f"제품명: {row['PRDLST_NM']}, 주된 기능성: {row['PRIMARY_FNCLTY']}, 원재료: {row['RAWMTRL_NM']}, 형태: {row['PRDT_SHAP_CD_NM']}",
    axis=1
).tolist()
embeddings = embedder.encode(text_data)

# 2. FAISS 인덱스에 임베딩 추가
dimension = embeddings.shape[1]  # 임베딩 벡터 차원
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# 3. 검색 함수
def search(query, k=3):
    query_embedding = embedder.encode([query])
    distances, indices = index.search(query_embedding, k)
    results = df_items.iloc[indices[0]]
    return results

# 4. GPT-3.5 Turbo를 이용한 RAG 시스템
def rag_qa_system(question):
    # 1) Retrieve: 유사한 데이터 검색
    search_results = search(question)

    # 2) Context 생성
    context = "\n".join(
        f"제품명: {row['PRDLST_NM']}, 주된 기능성: {row['PRIMARY_FNCLTY']}, 원재료: {row['RAWMTRL_NM']}, 형태: {row['PRDT_SHAP_CD_NM']}"
        for _, row in search_results.iterrows()
    )

    # 3) LLM에 질의와 컨텍스트 전달하여 답변 생성
    prompt = f"""
    당신은 건강 보조제 추천 전문가입니다. 사용자의 질문에 대해 아래 제공된 참고 정보에서 가장 관련이 있는 제품명을 찾아 추천해주세요.

    참고 정보:
    {context}

    질문: {question}

    답변 형식:
    가장 적합한 제품명과 해당 제품의 주요 원재료 및 기능성을 제시해주세요. 답변 예시:
    제품명: OOO
    주요 원재료: OOO
    기능성: OOO
    """

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are an assistant that provides specific supplement recommendations based on nutrients."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500
    )
    return response['choices'][0]['message']['content']

# 예제 질문
question = "눈이 안 좋을 때 도움이 되는 영양제를 찾아줘."
answer = rag_qa_system(question)
print(answer)
```
